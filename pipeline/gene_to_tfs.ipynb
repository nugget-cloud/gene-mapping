```python
import os
os.environ['PROTOCOL_BUFFERS_PYTHON_IMPLEMENTATION'] = 'python'

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from umap.umap_ import UMAP
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.feature_selection import SelectKBest, RFE
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans
# Load and preprocess data
data = pd.read_csv('/home/kunal/Desktop/reasearch/pipeline/Yeast_Tfs.csv')
data = data.drop('gene_id.2', axis=1)
data = data.set_index('gene_id.1')
data = data.drop_duplicates()

# Remove columns with non-numeric values
data = data.apply(pd.to_numeric, errors='coerce')
data = data.dropna(axis=1, how='any')

# Impute missing values
imputer = SimpleImputer(strategy='mean')
imputed_data = imputer.fit_transform(data)

data.head()
```

```python
#dimensionality reduction
reducer_tsne = TSNE(n_components=2, random_state=42)
reduced_data_tsne = reducer_tsne.fit_transform(data)

reducer_umap = UMAP(random_state=42)
reduced_data_umap = reducer_umap.fit_transform(data)
```

```python


# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(imputed_data, data.index, test_size=0.3, random_state=42)

# Feature selection using SelectKBest
k_best = SelectKBest(k=2)  # Select the top 2 features
X_train_k_best = k_best.fit_transform(X_train, y_train)
X_test_k_best = k_best.transform(X_test)

# Feature selection using Recursive Feature Elimination (RFE) with RandomForestClassifier
rfe = RFE(estimator=RandomForestClassifier(), n_features_to_select=2, step=1)
X_train_rfe = rfe.fit_transform(X_train, y_train)
X_test_rfe = rfe.transform(X_test)

```

```python

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Perform clustering using KMeans for t-SNE
n_clusters = 3
kmeans_tsne = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
cluster_assignments_tsne = kmeans_tsne.fit_predict(reduced_data_tsne)

# Perform clustering using KMeans for UMAP
kmeans_umap = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
cluster_assignments_umap = kmeans_umap.fit_predict(reduced_data_umap)

# Plot the t-SNE clustering assignments
plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
for i in range(n_clusters):
    plt.scatter(reduced_data_tsne[cluster_assignments_tsne == i, 0],
                reduced_data_tsne[cluster_assignments_tsne == i, 1],
                label=f'Cluster {i+1}')

plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.title('Clustering Assignments using t-SNE')

# Plot the UMAP clustering assignments
plt.subplot(1, 2, 2)
for i in range(n_clusters):
    plt.scatter(reduced_data_umap[cluster_assignments_umap == i, 0],
                reduced_data_umap[cluster_assignments_umap == i, 1],
                label=f'Cluster {i+1}')

plt.xlabel('UMAP Component 1')
plt.ylabel('UMAP Component 2')
plt.legend()
plt.title('Clustering Assignments using UMAP')

plt.show()
```
